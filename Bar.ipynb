```python
import matplotlib.pyplot as plt
import numpy as np


models = ['Perceptron', 'Logistic Regression', 'Decision Trees']
accuracy = [0.7229, 0.7359, 0.7056]
precision = [0.8537, 0.8000, 0.7986]
recall = [0.6954, 0.7947, 0.7351]
f1_score = [0.7664, 0.7973, 0.7655]


bar_width = 0.2


index = np.arange(len(models))


fig, ax = plt.subplots()


bar1 = ax.bar(index, accuracy, bar_width, label='Accuracy', align='center')
bar2 = ax.bar(index + bar_width, precision, bar_width, label='Precision', align='center')
bar3 = ax.bar(index + 2 * bar_width, recall, bar_width, label='Recall', align='center')
bar4 = ax.bar(index + 3 * bar_width, f1_score, bar_width, label='F1-Score', align='center')


ax.set_xlabel('Models')
ax.set_ylabel('Scores')
ax.set_title('Performance Metrics Comparison')
ax.set_xticks(index + 1.5 * bar_width)
ax.set_xticklabels(models)
ax.legend()


plt.tight_layout()
plt.show()
```
